```python
#importing pandas library
import pandas as pd
#reading csv file and assigning data to the variable
data = pd.read_csv("Salary_Data.csv")
data
```

```python
#Assigning independent and dependent variables respecitively
independent = data[["YearsExperience"]]
dependent = data[["Salary"]]
```

```python
#Scikit-learn provides utilities for splitting datasets into train and test sets.
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(independent,dependent,test_size=0.3 , random_state=0)
```

```python
x_train
```

```python
#This module within Scikit-learn contains implementations of various linear models for regression and classification tasks
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #create instance of the linear regression from sklearn library
regressor.fit(x_train,y_train) #fits the linear regression model to the training data
```

```python
weight = regressor.coef_ #weight of the training data
bias = regressor.intercept_ #bias of the training data
```

```python
weight
bias
```

```python
weight
```

```python
y_pred = regressor.predict(x_test) #model prediction values of the test dataset
```

```python
# computing the r2 score for test output and predicted output
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
#saving the model
import pickle 
filename = "finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#loadind the model
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#testing the model with the prediction
result = loaded_model.predict([[13]])
```
